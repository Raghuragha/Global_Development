### Handling Missing Values


```python
for col in df.select_dtypes(include=np.number):
    df[col].fillna(df[col].median(), inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

### 6.2 Correlation


```python
object_cols_to_convert = [
    'Business Tax Rate',
    'GDP',
    'Health Exp/Capita',
    'Tourism Inbound',
    'Tourism Outbound'
]

for col in object_cols_to_convert:
    if col in df.columns:
        # Convert to string to ensure string methods can be applied, handle NaNs.
        # Remove '$', ',', '%' and convert to numeric.
        df[col] = df[col].astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False).str.replace('%', '', regex=False)
        # Convert to numeric, coercing errors (e.g., 'nan' string) to NaN.
        df[col] = pd.to_numeric(df[col], errors='coerce')

# After cleaning and converting, fill any new NaNs that resulted from 'coerce' or were already present.
# This aligns with the previous median imputation strategy for numeric columns.
for col in object_cols_to_convert:
    if col in df.columns and df[col].isnull().any():
        # Avoid FutureWarning by assigning the result back instead of using inplace=True
        df[col] = df[col].fillna(df[col].median())

# Select only numeric columns for correlation calculation to avoid errors with 'Country' and other non-numeric columns.
numeric_df = df.select_dtypes(include=np.number)

plt.figure(figsize=(12,8))
sns.heatmap(numeric_df.corr(), cmap='coolwarm')
plt.show()
```

### 6.3 Outlier Detection


```python
for col in df.select_dtypes(include=np.number):
    sns.boxplot(x=df[col])
    plt.title(col)
    plt.show()
```

### Handling Outliers using IQR


```python
def cap_outliers(col):
    q1 = col.quantile(0.25)
    q3 = col.quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    return np.clip(col, lower, upper)

for col in df.select_dtypes(include=np.number):
    df[col] = cap_outliers(df[col])
```

## 🔷 7. Feature Engineering
### 7.1 Drop Non-Numeric Column


```python
df_model = df.drop(columns=['Country'])
```

### 7.2 Feature Scaling


```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_model)
df_scaled = pd.DataFrame(df_scaled, columns=df_model.columns)
```

### 7.3 Create Business Features


```python
# Development Index
df['Dev_Index'] = (df['GDP'] + df['Life Expectancy Female'] + df['Internet Usage']) / 3

# Risk Index
df['Risk_Index'] = (df['Infant Mortality Rate'] + df['Lending Interest']) / 2

# Business Score
df['Business_Score'] = (df['Ease of Business'] - df['Days to Start Business'] - df['Hours to do Tax'])
```

### Scale New Features

```python
new_features = ['Dev_Index', 'Risk_Index', 'Business_Score']
df[new_features] = scaler.fit_transform(df[new_features])
```

```python

import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

import matplotlib.pyplot as plt
import seaborn as sns
import joblib
```

```python
df_original = df.copy()

# Drop Country column for modeling
df = df.drop(columns=['Country'], errors='ignore')

# Fill missing values
df.fillna(df.mean(), inplace=True)

df.head()
```

```python

# Scaling
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
```

```python

# PCA
pca = PCA(n_components=0.95)
pca_data = pca.fit_transform(scaled_data)

print("Reduced Shape:", pca_data.shape)
```

```python

# Find optimal clusters
sil_scores = []
k_range = range(2, 10)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(pca_data)
    score = silhouette_score(pca_data, labels)
    sil_scores.append(score)
    print(f"K={k}, Score={score:.4f}")
```

```python

# Plot silhouette scores
plt.plot(k_range, sil_scores, marker='o')
plt.xlabel("Clusters")
plt.ylabel("Silhouette Score")
plt.title("Optimal K")
plt.show()
```

```python

# Best model
best_k = k_range[np.argmax(sil_scores)]
print("Best K:", best_k)

kmeans = KMeans(n_clusters=best_k, random_state=42, n_init=10)
clusters = kmeans.fit_predict(pca_data)

df_original['Cluster'] = clusters
```

```python

# Final evaluation
final_score = silhouette_score(pca_data, clusters)
print("Final Silhouette Score:", final_score)

df_original['Cluster'].value_counts()
```

```python

# Visualization
pca_2d = PCA(n_components=2)
reduced = pca_2d.fit_transform(scaled_data)

plt.scatter(reduced[:,0], reduced[:,1], c=clusters)
plt.xlabel("PCA1")
plt.ylabel("PCA2")
plt.title("Cluster Visualization")
plt.show()
```

```python

# Save models
joblib.dump(kmeans, "kmeans_model.joblib")
joblib.dump(scaler, "scaler.joblib")
joblib.dump(pca, "pca.joblib")

print("Models saved successfully!")
```
